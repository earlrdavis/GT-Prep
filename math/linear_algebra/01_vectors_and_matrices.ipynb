# Vectors & Matrices — Linear Algebra Warmup
*GT-Prep — Linear Algebra Notebook 01*

This notebook covers:
- Creating vectors and matrices in NumPy  
- Vector operations (magnitude, dot product)  
- Basic matrix multiplication  
- Geometric intuition  


```python
import numpy as np

# Create two simple vectors
v = np.array([1, 2, 3])
w = np.array([4, 5, 6])

v, w
```

```python
# Dot product
dot = np.dot(v, w)

# Magnitudes
mag_v = np.linalg.norm(v)
mag_w = np.linalg.norm(w)

dot, mag_v, mag_w
```

```python
# Simple matrix example
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

A, B
```

```python
# Matrix multiplication
C = np.matmul(A, B)
C
```

```python
import matplotlib.pyplot as plt

# Plot vector v
plt.figure(figsize=(5, 5))
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1)
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.title("Vector v in 2D")
plt.grid(True)
plt.show()
```
